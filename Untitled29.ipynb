```python
#feature enginerring
#conveting categorical data to numeric in column Type
#droping unnecessary columns
df_mod = df.drop(['step','nameOrig','nameDest','isFlaggedFraud'],axis = 1)
df_mod.head(5)
```

```python
df_mod['type'].unique()
```

```python
df_mod['type'] = df_mod['type'].replace(['PAYMENT','TRANSFER','CASH_OUT','DEBIT','CASH_IN'],[1,2,3,4,5])
```

```python
df_mod['type'].unique()
```

```python
df_mod.columns
```

```python
#assigning the target variable isFraud
X = df_mod.drop('isFraud',axis = 1)
y = df_mod['isFraud']
```

```python
#spliting data to training and testing dataets
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size= 0.3)
```

```python
#feature Scaling

sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
# initiate the regression model

model_lr = LogisticRegression()
model_lr.fit(X_train_sc,y_train)


```

```python
y_pred = model_lr.predict(X_test_sc)
```

```python
y_pred
```

```python
y_test
```

```python
# accuracy score is used to check how many are wrong predictions and right predictions 
```

```python
accuracy_score(y_test,y_pred) * 100
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
